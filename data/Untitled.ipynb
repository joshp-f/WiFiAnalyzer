```python
import pandas as pd
```

```python
df = pd.read_csv('log1.txt', sep='|')
df2 = pd.read_csv('log2.txt', sep='|')
```

```python
def preprocess(df):
    df = df.drop(df[df['Time Stamp'] =='Time Stamp'].index)
    df['Time Stamp'] = pd.to_datetime(df['Time Stamp']) 
    df = df.reset_index()
    return df
def generateBSSIDgroups(df):
    d = df.groupby(by=['Time Stamp']).groups

    data = {}
    for k,l in d.items():
        data[str(k)] =  set(df['BSSID'].iloc[l])
    return data

def getTimestampPairs(df1,df2):
    df1 = df1.drop_duplicates(subset=['Time Stamp'])
    df2 = df2.drop_duplicates(subset=['Time Stamp'])
    df2['Alt Time Stamp'] = df2['Time Stamp']
    
    cdf = pd.merge_asof(df1,df2,on='Time Stamp',tolerance = pd.Timedelta("16 minutes"), direction = "nearest")
    pairs = cdf[['Time Stamp', 'Alt Time Stamp']]
    return pairs
```

```python
df1 = preprocess(df)
df2 = preprocess(df2)
```

```python
BSSID1 = generateBSSIDgroups(df1)
BSSID2 = generateBSSIDgroups(df2)
print(BSSID1)
```

```python
pairs = getTimestampPairs(df1, df2)
print(pairs)
```

```python
for r in range(len(pairs)):
    a,b = list(map(lambda v : str(pd.Timestamp(v)), pairs.iloc[r].values))
    print(sorted(list(BSSID1[a])),sorted(list(BSSID2[b])))
    print(len(BSSID1[a] & BSSID2[b]) / len(BSSID1[a] | BSSID2[b]))
```
